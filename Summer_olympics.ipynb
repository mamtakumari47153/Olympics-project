# Summer Olympics Data Analysis Assignment

# 1. In how many cities Summer Olympics is held so far?


```python
import pandas as pd

df = pd.read_csv('summer.csv')
```

```python
data = []

for City in df['City'].unique():
    data.append((City, len(df[df['City']== City])))
    
data = pd.DataFrame(data, columns=['City', 'Frequency'])
print('\n**Data Representation** ')
print(data)
```

```python
data = []

for city in df['City'].unique():
    data.append([city , len(df[df['City']  == city])])
# print(data)
pd.DataFrame(data,columns = ['City','Frequency']).sort_values(by='Frequency', ascending=False).plot(x = 'City', y = 'Frequency', kind = 'bar', figsize = (10,5))
print("**data representation in graph**")
```

# 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[(df['Sport']  == sport) & (df['Medal'] == "Gold")])])
#print(data)
pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (5,5))
print("!! most number of gold medals!!")
```

# 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []
for model in df['Sport'].unique():
    data.append((model, len(df[df['Sport'] == model])))
    
data = pd.DataFrame(data, columns = ['Sports', 'Medals'])

data = data.sort_values(by = 'Medals', ascending = False).head()
print('**** Top-5 sports having most number of Medals ****\n')
print(data)

data.plot(x = 'Sports', y = 'Medals', kind = 'line')
print('\nData Representation of Top-5 sports having most number of Medals using line-Graph!! ')
```

# 4. Which player has won most number of medals? (Top 5)

```python
data = []

for athlete in df['Athlete'].unique():
    data.append([athlete , len(df[df['Athlete'] == athlete])])
# print(data)
pd.DataFrame(data,columns = ['Athlete','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (5,5))
print("****\ndata representation which player has won most number of medals****")
```

# 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
dc = df[df['Medal'] == 'Gold']
data = []
for model in dc['Athlete'].unique():
    data.append((model, len(dc[dc['Athlete'] == model])))
    
data = pd.DataFrame(data, columns = ['Athlete', 'Gold_Medals'])

data = data.sort_values(by = 'Gold_Medals', ascending = False).head()
print('**** Top-5 Players having most number of Gold_Medals ****\n')
print(data)

data.plot(x = 'Athlete', y = 'Gold_Medals', kind = 'bar', color = 'red')
print('\nData Representation of Top-5 Players having most number of Gold Medals using bar-Graph!! ')
```

# 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = []

for year in df['Year'].unique():
    data.append([year , len(df[(df['Year'] == year) & (df['Medal'] == "Gold") & (df['Country'] == "IND")])])
print("India Got First Gold Medal In 1928 ")
pd.DataFrame(data,columns = ['Year','Gold']).sort_values(by='Year', ascending=True).plot(x = 'Year', y = 'Gold', kind = 'bar', figsize = (10,5))
print("****won first gold medal in summer olympics***")
```

# 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for event in df['Event'].unique():
    data.append([event , len(df[df['Event'] == event])])

# print(data)
pd.DataFrame(data,columns = ['Event','No. of Athlete']).sort_values(by='No. of Athlete', ascending=False).head().plot(x = 'Event', y = 'No. of Athlete', kind = 'bar', figsize = (5,5))
print("data representation")
```

# 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[(df['Sport'] == sport) & (df['Medal'] == "Gold") & (df['Gender'] == "Women")])])
# print(data)
pd.DataFrame(data,columns = ['Sport','Female Gold Medalists']).sort_values(by='Female Gold Medalists', ascending=False).head().plot(x = 'Sport', y = 'Female Gold Medalists', kind = 'bar', figsize = (5,5))
print("**Gold medalist**")
```
